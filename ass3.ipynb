```python
x = data[['Time_on_App','Length_of_Membership']]
y = data['Yearly_Amount_Spent']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.25,random_state = 85)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
model1 = lm.fit(X_train,Y_train)
```

```python
#print intrcept and coefficients
b0=lm.intercept_
b1=lm.coef_
print(b0,b1)
```

```python
y_predict=lm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('RSquared:',metrics.r2_score(Y_test,y_predict))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(Y_test,y_predict)))
```

```python
plt.scatter(Y_test,y_predict)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```
